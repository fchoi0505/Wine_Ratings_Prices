Machine Learning: Continue working with 2017 dataframe

```python
# Create a dataframe with country, price, points, and variety
new2017_df = dropped_df2017[['country', 'variety','price','points']]
new2017_df.head()
```

```python
#rename columns in dataframe to be more fitting
new2017_df = new2017_df.rename(columns = {'points':'ratings_points','price':'price_dollars','variety':'wine_type'},inplace=False)
new2017_df.head()
```

Create a categorical value
convert rating points into good_wine or great_wine based on ratings: http://www.winewins.com/wine-ratings/
good_wine: rating 89 or below
great_wine: 90 or above

```python
# Convert the TARGET column values to low_risk and high_risk based on their values
new2017_df.loc[new2017_df.ratings_points < 90 , "ratings_desc"] = 0 #good_wine
```

```python
new2017_df.loc[new2017_df.ratings_points > 89 , "ratings_desc"] = 1 #great_wine
```

```python
# validate the ratings_desc column is working
new2017_df.sort_values(['ratings_points'], ascending=True)
```

## Remove Outliers

```python
#Remove outliers based on price drop versus inclusion 3 standard deviations
minus_outliers_df = new2017_df[new2017_df['price_dollars'] < 100]

sns.distplot(minus_outliers_df['price_dollars'])
```

Final Dataframe for Machine learning
Choose to pass : (1) a single country -or- (2) all countries

```python
#testing with all countries
machine_df = minus_outliers_df[['price_dollars','ratings_desc']]
machine_df.head()
```

```python
# Create our features
X = machine_df.drop('ratings_desc', axis =1 )

# Create our target
y = machine_df['ratings_desc']
```

```python
#Add Standard Scaling to further optimize model findings
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)


#Normalizing the features
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

```python
# Train the data
classifier.fit(X_train, y_train)
```

```python
# Predict outcomes for test data set
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

## Validate the model using the test data with changed outlier removal and standard scaling applied

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
#Removing outliers for price >$100 and adding standard scaling had negligible impact to the model
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(predictions,y_test)
cm
```

```python
# Format for easier understanding
cm_df = pd.DataFrame(cm)
cm_df.columns = ['Predicted good_wine','Predicted great_wine']
cm_df = cm_df.rename(index={0: 'Actual good_wine',1:'Actual great_wine'})
cm_df
```

```python
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test,predictions))
```
